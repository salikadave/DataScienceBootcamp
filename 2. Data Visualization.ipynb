### scatter() plots points on the graph
#### marker is the type of the point (default is circle)

```python
X = [590,540,740,130,810,300,320,230,470,620,770,250]  #no. of items sold
Y = [32,36,39,52,61,72,77,75,68,57,48,48] # temp in fahrenheit

#Syntax: plt.scatter(X,Y)
plt.scatter(X,Y, s=60, c='red', marker='^')

plt.title("Relationship between Temperature and Cold Coffee Sales")
plt.xlabel("Cups of Iced Coffee Sold")
plt.ylabel("Temperature in Fahrenheit")
plt.show()  #works even without this
```

## Example 2
### Days v/s min-max temp in celsius

```python
celsius_min = []  #[19.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1] 
celsius_max = []  #[24.8, 28.9, 31.3, 33.0, 34.9, 35.6, 38.4, 39.2]
num = int(input("Enter the number of days: "))
days = list((range(1,num+1)))
print(days)

for i in range(0,num):
    min = float(input("Enter minimum: "))
    celsius_min.append(min)
    max = float(input("Enter maximum: "))
    celsius_max.append(max)
    
print(celsius_min)
print(celsius_max)

plt.xlabel("Day")
plt.ylabel("Temperature in Degrees Celsius")
plt.plot(days, celsius_min,
         days, celsius_min, "sg",
         days, celsius_max,
         days, celsius_max, "or")

#changing the range of the axis
plt.ylim(-20,100)

plt.show()
```

# 3D Plots

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4,4,0.25)
Y = np.arange(-4,4,0.25)
X, Y = np.meshgrid(X,Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=3, cstride=3, cmap=plt.cm.hot)
ax.contourf(X, Y, Z, zdir='z', offset=-2, cmap=plt.cm.hot)
ax.set_zlim(-2,2)

plt.show()
```

# Seaborn

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline 
#use this when there is a GUI involved
import seaborn as sns
```

```python
dataFrame = pd.read_csv('Pokemon2.csv',index_col=0)
dataFrame.head()
```

```python
sns.lmplot(x='Attack', y='Defense', data=dataFrame)
```

```python
sns.lmplot(x='Attack', y='Defense', data=dataFrame,
           fit_reg=False,  #No regression line
           hue='Stage')   #Color by evolution stage
```

```python
sns.boxplot(data=dataFrame)
```

```python
stats_df = dataFrame.drop(['Total','Stage','Legendary'], axis=1)
sns.boxplot(data=stats_df)
```
